Use MAD, MAPE and RMSE to compare these three models. Which model is the best predictor based on these metrics? (You can use the Markdown lines to answer verbal questions).

```python
# There are Three Module used MAS, MAPE and RMSE following actual values and predications ina Pandas DataFrame
# I have Pandas Library installed
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('/content/CSV File for Evaluation Metrics Exercise, Module 3-1.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Calculate Mean Absolute Deviation (MAD) for each model
mad_model1 = np.mean(np.abs(df['Actual Y'] - df['Model 1 Prediction of Y']))
mad_model2 = np.mean(np.abs(df['Actual Y'] - df['Model 2 Prediction of Y']))
mad_model3 = np.mean(np.abs(df['Actual Y'] - df['Model 3 Prediction of Y']))
```

```python
# Print the evaluation metrics for each model
print("Model 1:")
print("MAD:", mad_model1)
```

```python
# Calculate Mean Absolute Percentage Error (MAPE) for each model
mape_model1 = np.mean(np.abs((df['Actual Y'] - df['Model 1 Prediction of Y']) / df['Actual Y'])) * 100
mape_model2 = np.mean(np.abs((df['Actual Y'] - df['Model 2 Prediction of Y']) / df['Actual Y'])) * 100
mape_model3 = np.mean(np.abs((df['Actual Y'] - df['Model 3 Prediction of Y']) / df['Actual Y'])) * 100
```

```python
print("\nModel 2:")
print("MAD:", mad_model2)
print("MAPE:", mape_model2)
```

```python
# Calculate Root Mean Squared Error (RMSE) for each model
rmse_model1 = np.sqrt(mean_squared_error(df['Actual Y'], df['Model 1 Prediction of Y']))
rmse_model2 = np.sqrt(mean_squared_error(df['Actual Y'], df['Model 2 Prediction of Y']))
rmse_model3 = np.sqrt(mean_squared_error(df['Actual Y'], df['Model 3 Prediction of Y']))
```

```python
print("\nModel 3:")
print("MAD:", mad_model3)
print("MAPE:", mape_model3)
print("RMSE:", rmse_model3)
```

```python
# Print the results of Models:
print("Model 1 - MAD:", mad_model1, "MAPE:", mape_model1, "RMSE:", rmse_model1)
print("Model 2 - MAD:", mad_model2, "MAPE:", mape_model2, "RMSE:", rmse_model2)
print("Model 3 - MAD:", mad_model3, "MAPE:", mape_model3, "RMSE:", rmse_model3)
```

# The best model based on RMSE is Model 2

```python
# Results
# Create a DataFrame to display the metrics
metrics_df = pd.DataFrame({
    'Model': ['Model 1', 'Model 2', 'Model 3'],
    'MAD': [mad_model1, mad_model2, mad_model3],
    'MAPE': [mape_model1, mape_model2, mape_model3],
    'RMSE': [rmse_model1, rmse_model2, rmse_model3]
})

# Print the metrics DataFrame
print(metrics_df)

# Determine the best model based on the metrics
best_model = metrics_df.loc[metrics_df['RMSE'].idxmin()]['Model']
print(f"The best model based on RMSE is: {best_model}")
```
